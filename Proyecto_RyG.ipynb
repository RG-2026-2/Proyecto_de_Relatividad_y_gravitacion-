# Fundamentos Teóricos: Simulación Newtoniana de un Disco de Acreción

##  Introducción Física

Un disco de acreción es una estructura formada por gas y polvo que orbita alrededor de un objeto central masivo, como un agujero negro, una estrella de neutrones o una estrella joven. La materia que cae hacia el objeto central pierde momento angular debido a fuerzas viscosas, se calienta por fricción y emite radiación electromagnética en un amplio rango de longitudes de onda.

En este proyecto, nos enfocamos en la **dinámica orbital newtoniana** como primera aproximación, antes de incorporar efectos relativistas. Aunque esta aproximación falla cerca del horizonte de eventos de un agujero negro, es útil para:
- Comprender la estructura básica del disco
- Establecer un punto de comparación con modelos más complejos
- Desarrollar las herramientas computacionales necesarias

## Potencial Gravitacional Newtoniano

Para un agujero negro tratado como una masa puntual $M$, el potencial gravitacional en el vacío es:

$$
\Phi(r) = -\frac{GM}{r}
$$

donde:
- $G = 6.674 \times 10^{-11} \, \text{m}^3 \text{kg}^{-1} \text{s}^{-2}$ es la constante gravitacional
- $r$ es la distancia al centro

La fuerza gravitacional sobre una partícula de masa $m$ (despreciable frente a $M$) es:

$$
\mathbf{F} = -m \nabla \Phi = -\frac{GMm}{r^2} \hat{\mathbf{r}}
$$

## Ecuaciones de Movimiento

La segunda ley de Newton para una partícula de prueba (masa despreciable) en el campo gravitacional del agujero negro es:

$$
\frac{d^2 \mathbf{r}}{dt^2} = -\frac{GM}{r^3} \mathbf{r}
$$

En coordenadas cartesianas $(x, y, z)$, esto se descompone en tres ecuaciones diferenciales ordinarias acopladas:

$$
\begin{aligned}
\ddot{x} &= -\frac{GM}{(x^2 + y^2 + z^2)^{3/2}} \, x \\
\ddot{y} &= -\frac{GM}{(x^2 + y^2 + z^2)^{3/2}} \, y \\
\ddot{z} &= -\frac{GM}{(x^2 + y^2 + z^2)^{3/2}} \, z
\end{aligned}
$$

## Cantidades Conservadas (Simetrías)

### Energía Específica

El potencial newtoniano es independiente del tiempo, por lo que la energía mecánica se conserva:

$$
E = \frac{1}{2} v^2 - \frac{GM}{r} = \text{constante}
$$

### Momento Angular Específico

El potencial es esféricamente simétrico (solo depende de $r$), por lo que el momento angular se conserva:

$$
\mathbf{L} = \mathbf{r} \times \mathbf{v} = \text{constante}
$$

En el plano ecuatorial ($z=0$), el momento angular tiene solo componente $z$:

$$
L_z = x v_y - y v_x = \text{constante}
$$

## Órbitas Circulares Keplerianas

Para una órbita circular de radio $r$, la aceleración centrípeta es proporcionada por la gravedad:

$$
\frac{v^2}{r} = \frac{GM}{r^2} \quad \Rightarrow \quad v_{\text{circ}} = \sqrt{\frac{GM}{r}}
$$

El período orbital es:

$$
T = \frac{2\pi r}{v_{\text{circ}}} = 2\pi \sqrt{\frac{r^3}{GM}}
$$

La energía específica de una órbita circular es:

$$
E_{\text{circ}} = \frac{1}{2} v_{\text{circ}}^2 - \frac{GM}{r} = \frac{GM}{2r} - \frac{GM}{r} = -\frac{GM}{2r}
$$

## Órbitas Elípticas (Kepler)

La solución general de la ecuación de movimiento es una cónica (elipse, parábola o hipérbola). Para órbitas ligadas (elipses), la ecuación de la trayectoria en coordenadas polares $(r, \phi)$ es:

$$
r(\phi) = \frac{a(1 - e^2)}{1 + e \cos(\phi - \phi_0)}
$$

donde:
- $a$ es el semieje mayor
- $e$ es la excentricidad ($0 \le e < 1$)
- $\phi_0$ es el ángulo del periastro

## El Disco de Acreción Newtoniano

### Modelo Simplificado

En nuestra simulación, el disco se aproxima como un conjunto de **partículas de prueba** que:
- No interactúan entre sí (disco no colisional)
- Orbitan independientemente alrededor del agujero negro
- Tienen masas despreciables frente a $M$

### Distribución Radial

Para modelar un disco con **densidad superficial constante** $\Sigma_0$ (número de partículas por unidad de área), la distribución radial debe satisfacer:

$$
\frac{dN}{dr} \propto r
$$

Integrando, obtenemos que el radio debe muestrearse como:

$$
r = \sqrt{r_{\text{min}}^2 + u \left(r_{\text{max}}^2 - r_{\text{min}}^2\right)}, \quad u \sim \text{Uniform}(0,1)
$$

### Distribución de Velocidades

En una primera aproximación, las partículas tienen velocidades puramente tangenciales (órbitas circulares). Para simular la **temperatura** del disco (dispersión de velocidades debida a turbulencia o colisiones), añadimos una pequeña perturbación aleatoria:

$$
v = v_{\text{circ}} \cdot (1 + \varepsilon), \quad \varepsilon \sim \mathcal{N}(0, \sigma_v)
$$

donde $\sigma_v$ es la dispersión fraccional (típicamente $0.01$ a $0.05$).

## Unidades Adimensionales

Para simplificar los cálculos numéricos, trabajamos en **unidades adimensionales** donde:

$$
G = 1, \quad M = 1
$$

En estas unidades:
- La unidad de distancia es arbitraria (puede escalarse después)
- La unidad de tiempo es tal que $v_{\text{circ}} = 1/\sqrt{r}$
- El período orbital a $r = 1$ es $2\pi$

**Conversión a unidades físicas:** Si se desea escalar a un sistema real (ej: un agujero negro de $10 M_{\odot}$), se aplica:

$$
r_{\text{físico}} = r_{\text{adim}} \cdot r_0, \quad t_{\text{físico}} = t_{\text{adim}} \cdot t_0
$$

con $r_0 = GM/c^2$ (radio gravitacional) o cualquier otra escala deseada.

## Integración Numérica

### Integradores Simplécticos

Para sistemas hamiltonianos como este, los **integradores simplécticos** (como WHFast) conservan la energía y el momento angular mucho mejor a largo plazo que métodos como Runge-Kutta.

**WHFast (Wisdom-Holman):** Específicamente diseñado para problemas de N-cuerpos con una masa central dominante.

### Paso de Tiempo

El paso de tiempo $\Delta t$ debe ser pequeño comparado con el período orbital más corto:

$$
\Delta t \ll T_{\text{min}} = 2\pi \sqrt{\frac{r_{\text{min}}^3}{GM}}
$$

En nuestras unidades, para $r_{\text{min}} = 1$, $T_{\text{min}} = 2\pi \approx 6.28$. Elegimos $\Delta t = 0.01$, que es $\sim 0.16\%$ del período mínimo.


```python
import numpy as np
import rebound
import plotly.graph_objects as go
from typing import Tuple, Optional
```

```python
def crear_simulacion_disco_newtoniano(
    n_particulas: int = 1000,
    masa_bh: float = 1.0,
    r_min: float = 1.0,
    r_max: float = 6.0,
    dispersion_vel: float = 0.03,
    dt: float = 0.01,
    semilla: int = 42,
) -> rebound.Simulation:
    """
    Crea una simulacion rebound de un disco de particulas de prueba.

    La distribucion radial de las particulas se elige para que la densidad
    superficial (numero de particulas por unidad de area) sea aproximadamente
    constante. Esto se logra muestreando r^2 uniformemente.

    Las velocidades iniciales son aproximadamente circulares (keplerianas)
    con una pequena dispersion aleatoria que simula la "temperatura" del disco.

    Parametros
    ----------
    n_particulas : int
        Numero de particulas en el disco (excluyendo el agujero negro).
    masa_bh : float
        Masa del agujero negro central (en unidades adimensionales).
    r_min : float
        Radio interior del disco (en unidades adimensionales).
    r_max : float
        Radio exterior del disco (en unidades adimensionales).
    dispersion_vel : float
        Dispersion fraccional de las velocidades (desviacion estandar relativa).
        Ej: 0.03 significa 3% de dispersion gaussiana.
    dt : float
        Paso de tiempo de integracion (en unidades adimensionales).
    semilla : int
        Semilla para el generador de numeros aleatorios (reproducibilidad).

    Retorna
    -------
    rebound.Simulation
        Objeto de simulacion de rebound listo para integrar.
    """

    # =========================================================================
    # Configuracion basica de la simulacion
    # =========================================================================

    # Crear objeto de simulacion vacio
    sim = rebound.Simulation()

    # Constante gravitacional (en nuestras unidades, G = 1)
    sim.G = 1.0

    # Integrador WHFast (Wisdom-Holman):
    #   - Es un integrador simpléctico de orden superior
    #   - Ideal para sistemas con una masa central dominante
    #   - Conserva la energia a largo plazo mucho mejor que Runge-Kutta
    sim.integrator = "whfast"

    # Paso de tiempo fijo (WHFast funciona mejor con paso constante)
    sim.dt = dt

    # =========================================================================
    # Agujero negro central (cuerpo primario)
    # =========================================================================

    # El agujero negro se coloca en el origen (x=y=z=0) en reposo (vx=vy=vz=0)
    # El parametro 'hash' permite identificar este cuerpo facilmente despues
    sim.add(m=masa_bh, hash="BH")

    # =========================================================================
    # Generacion de particulas del disco
    # =========================================================================

    # Generador de numeros aleatorios con semilla fija (reproducibilidad)
    rng = np.random.default_rng(semilla)

    # Precalcular valores constantes para optimizacion
    r_min2 = r_min * r_min   # r_min^2
    r_max2 = r_max * r_max   # r_max^2

    # Bucle de creacion de particulas
    for i in range(n_particulas):

        # -----------------------------------------------------------------
        # Posicion radial (distribucion de densidad superficial constante)
        # -----------------------------------------------------------------
        # Para que el numero de particulas por unidad de area sea constante,
        # debemos muestrear r^2 uniformemente entre r_min^2 y r_max^2.
        # Justificacion: El area de un anillo es dA = 2π r dr.
        # Si queremos dN/dA = constante, entonces dN/dr ∝ r.
        # Integrando: N(r) ∝ r^2, por lo que r ∝ sqrt(N).
        # Por lo tanto, tomamos u ~ Uniform(0,1) y hacemos r = sqrt(r_min^2 + u*(r_max^2 - r_min^2))
        # -----------------------------------------------------------------
        u = rng.uniform(0.0, 1.0)
        r = np.sqrt(r_min2 + u * (r_max2 - r_min2))

        # Angulo azimutal (distribucion uniforme entre 0 y 2π)
        theta = rng.uniform(0.0, 2.0 * np.pi)

        # Coordenadas cartesianas
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        # Disco en el plano ecuatorial (z = 0)
        z = 0.0

        # -----------------------------------------------------------------
        # Velocidad orbital (circular kepleriana con dispersion)
        # -----------------------------------------------------------------
        # En unidades adimensionales (G=1, M_BH=1):
        #   v_circ = sqrt(G*M_BH / r) = 1/√r
        v_circ = np.sqrt(sim.G * masa_bh / r)

        # Añadir dispersion gaussiana a la velocidad
        #   factor = 1 + ε, donde ε ~ N(0, dispersion_vel)
        #   Usamos max(0, ...) para evitar velocidades negativas (muy improbable)
        factor = 1.0 + rng.normal(0.0, dispersion_vel)
        v = max(0.0, v_circ * factor)

        # Velocidad tangencial en coordenadas cartesianas:
        #   El vector velocidad para una orbita circular es perpendicular
        #   al vector posicion: v = v * (-sinθ, cosθ, 0)
        vx = -v * np.sin(theta)
        vy = v * np.cos(theta)
        vz = 0.0

        # -----------------------------------------------------------------
        # Añadir particula a la simulacion
        # -----------------------------------------------------------------
        # m = 0.0 indica que es una particula de prueba:
        #   - No ejerce fuerza gravitacional sobre otros cuerpos
        #   - Solo es afectada por la gravedad del agujero negro
        # Esto es valido si la masa del disco es despreciable frente a M_BH
        sim.add(m=0.0, x=x, y=y, z=z, vx=vx, vy=vy, vz=vz, hash=f"p{i}")

    # =========================================================================
    # Ajuste final: centrar el sistema en el centro de masa
    # =========================================================================
    # Mueve el origen de coordenadas al centro de masa del sistema.
    # Dado que solo el agujero negro tiene masa, esto simplemente garantiza
    # que el BH quede exactamente en el origen (si ya lo estaba, no cambia nada).
    sim.move_to_com()

    return sim
```

```python
def integrar_y_guardar(
    sim: rebound.Simulation,
    tiempo_total: float,
    n_pasos: int,
    verbose: bool = True,
) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
    """
    Integra la simulacion y guarda las trayectorias y velocidades.

    Parametros
    ----------
    sim : rebound.Simulation
        Simulacion a integrar.
    tiempo_total : float
        Tiempo total de integracion (en unidades adimensionales).
    n_pasos : int
        Numero de pasos de tiempo en los que se guardaran los datos.
    verbose : bool
        Si es True, imprime informacion de progreso.

    Retorna
    -------
    tuple
        (tiempos, posiciones, velocidades, n_particulas)
        - tiempos : array de tiempos (shape: n_pasos)
        - posiciones : array de posiciones (x,y) (shape: n_pasos, n_particulas, 2)
        - velocidades : array de velocidades (vx,vy) (shape: n_pasos, n_particulas, 2)
        - n_particulas : numero de particulas (excluyendo el BH)
    """

    # Excluir el agujero negro central de las particulas a guardar
    # sim.particles[0] es el BH, sim.particles[1:] son las particulas del disco
    n_particulas = sim.N - 1

    # Crear array de tiempos
    tiempos = np.linspace(0.0, tiempo_total, n_pasos)

    # Inicializar arrays para guardar datos
    # posiciones[k, j, 0] = x de particula j en paso k
    # posiciones[k, j, 1] = y de particula j en paso k
    posiciones = np.zeros((n_pasos, n_particulas, 2))

    # velocidades[k, j, 0] = vx de particula j en paso k
    # velocidades[k, j, 1] = vy de particula j en paso k
    velocidades = np.zeros((n_pasos, n_particulas, 2))

    if verbose:
        print(f"Iniciando integracion: {n_particulas} particulas, {n_pasos} pasos")
        print(f"Tiempo total: {tiempo_total:.2f}, dt = {sim.dt:.4f}")

    # Bucle principal de integracion
    for k, t in enumerate(tiempos):

        # Integrar hasta el tiempo t
        # exact_finish_time=0 significa "no ajustes de paso"
        # Esto asegura que el integrador llegue exactamente al tiempo pedido
        sim.integrate(t, exact_finish_time=0)

        # Guardar estado de cada particula
        for j in range(n_particulas):
            # La particula j+1 (indice 0 es el BH)
            p = sim.particles[j + 1]

            # Posiciones
            posiciones[k, j, 0] = p.x
            posiciones[k, j, 1] = p.y

            # Velocidades
            velocidades[k, j, 0] = p.vx
            velocidades[k, j, 1] = p.vy

        # Reporte de progreso (cada 10% de avance)
        if verbose and (k + 1) % max(1, n_pasos // 10) == 0:
            progreso = 100.0 * (k + 1) / n_pasos
            print(f"  Progreso: {progreso:.0f}%")

    if verbose:
        print("Integracion completada.")

    return tiempos, posiciones, velocidades, n_particulas
```

```python
def graficar_trayectorias(
    posiciones: np.ndarray,
    n_muestra: int = 60,
    titulo: str = "Disco de particulas en potencial newtoniano",
    guardar_figura: Optional[str] = None,
 ) -> None:
    """
    Grafica las trayectorias de una muestra de particulas y el estado final.

    Parametros
    ----------
    posiciones : np.ndarray
        Array de posiciones con forma (n_pasos, n_particulas, 2).
    n_muestra : int
        Numero de particulas a mostrar en las trayectorias.
    titulo : str
        Titulo del grafico.
    guardar_figura : str, optional
        Si se proporciona, guarda la figura en esa ruta.
    """
    n_pasos, n_particulas, _ = posiciones.shape

    # Seleccionar una muestra de particulas para graficar trayectorias
    # Usamos indices espaciados uniformemente para una buena cobertura radial
    indices_muestra = np.linspace(0, n_particulas - 1, n_muestra, dtype=int)

    fig = go.Figure()

    # -----------------------------------------------------------------
    # Graficar trayectorias completas de la muestra (3D con z=0)
    # -----------------------------------------------------------------
    for idx in indices_muestra:
        fig.add_trace(
            go.Scatter3d(
                x=posiciones[:, idx, 0],
                y=posiciones[:, idx, 1],
                z=np.zeros(n_pasos),
                mode="lines",
                line=dict(width=2),
                opacity=0.6,
                showlegend=False,
                hoverinfo="skip",
            )
        )

    # -----------------------------------------------------------------
    # Marcar el agujero negro central
    # -----------------------------------------------------------------
    fig.add_trace(
        go.Scatter3d(
            x=[0.0],
            y=[0.0],
            z=[0.0],
            mode="markers",
            marker=dict(size=6, color="black"),
            name="Agujero negro",
        )
    )

    # -----------------------------------------------------------------
    # Marcar la posicion final de TODAS las particulas
    # -----------------------------------------------------------------
    fig.add_trace(
        go.Scatter3d(
            x=posiciones[-1, :, 0],
            y=posiciones[-1, :, 1],
            z=np.zeros(n_particulas),
            mode="markers",
            marker=dict(size=2, opacity=0.5),
            name="Particulas (estado final)",
        )
    )

    # -----------------------------------------------------------------
    # Ajustes esteticos del grafico
    # -----------------------------------------------------------------
    fig.update_layout(
        title=titulo,
        template="plotly_white",
        width=800,
        height=700,
        margin=dict(l=40, r=40, t=60, b=40),
        scene=dict(
            xaxis_title="x",
            yaxis_title="y",
            zaxis_title="z",
            aspectmode="data",
        ),
    )

    if guardar_figura:
        if guardar_figura.lower().endswith(".html"):
            fig.write_html(guardar_figura)
            print(f"Figura guardada en: {guardar_figura}")
        else:
            try:
                fig.write_image(guardar_figura)
                print(f"Figura guardada en: {guardar_figura}")
            except Exception as exc:
                html_path = guardar_figura + ".html"
                fig.write_html(html_path)
                print(
                    "No se pudo guardar la imagen; se guardo HTML en: "
                    f"{html_path}. Error: {exc}"
                )

    fig.show()
```

```python
# =============================================================================
# EJECUCION PRINCIPAL
# =============================================================================

if __name__ == "__main__":

    # -------------------------------------------------------------------------
    # Parametros de la simulacion
    # -------------------------------------------------------------------------
    # Estos valores se pueden modificar para explorar diferentes configuraciones
    PARAMETROS = {
        "n_particulas": 1000,      # numero de particulas en el disco
        "masa_bh": 1.0,            # masa del agujero negro (adimensional)
        "r_min": 1.0,              # radio interior del disco
        "r_max": 6.0,              # radio exterior del disco
        "dispersion_vel": 0.03,    # dispersion fraccional de velocidades (3%)
        "dt": 0.01,                # paso de tiempo de integracion
        "semilla": 42,             # semilla para reproducibilidad
    }

    # Parametros de integracion
    TIEMPO_TOTAL = 120.0           # tiempo total de simulacion
    N_PASOS = 1000                 # numero de snapshots a guardar

    # -------------------------------------------------------------------------
    # Crear la simulacion
    # -------------------------------------------------------------------------
    print("=" * 60)
    print("SIMULACION NEWTONIANA DE DISCO DE ACRECION")
    print("=" * 60)
    print("\nCreando simulacion...")
    print(f"  - {PARAMETROS['n_particulas']} particulas")
    print(f"  - r ∈ [{PARAMETROS['r_min']}, {PARAMETROS['r_max']}]")
    print(f"  - Dispersion de velocidades: {PARAMETROS['dispersion_vel']*100:.1f}%")
    print(f"  - Paso de tiempo dt = {PARAMETROS['dt']}")

    sim = crear_simulacion_disco_newtoniano(**PARAMETROS)

    # -------------------------------------------------------------------------
    # Integrar y guardar datos
    # -------------------------------------------------------------------------
    print("\nIntegrando...")
    tiempos, posiciones, velocidades, n_part = integrar_y_guardar(
        sim=sim,
        tiempo_total=TIEMPO_TOTAL,
        n_pasos=N_PASOS,
        verbose=True,
    )

    # -------------------------------------------------------------------------
    # Graficar resultados
    # -------------------------------------------------------------------------
    print("\nGenerando graficos...")
    graficar_trayectorias(
        posiciones=posiciones,
        n_muestra=60,
        titulo="Disco de acrecion - Potencial Newtoniano",
        # guardar_figura="disco_newtoniano.png",  # Descomentar para guardar
    )

    # -------------------------------------------------------------------------
    # Informacion estadistica final
    # -------------------------------------------------------------------------
    # Calcular radios finales para ver la distribucion
    radios_finales = np.sqrt(posiciones[-1, :, 0]**2 + posiciones[-1, :, 1]**2)

    print("\n" + "-" * 40)
    print("ESTADISTICAS FINALES")
    print("-" * 40)
    print(f"Radio medio final: {np.mean(radios_finales):.3f}")
    print(f"Desv. est. radio final: {np.std(radios_finales):.3f}")
    print(f"Radio minimo final: {np.min(radios_finales):.3f}")
    print(f"Radio maximo final: {np.max(radios_finales):.3f}")

    # Verificar si alguna particula colisiono con el BH (r < r_min)
    colisionaron = radios_finales < PARAMETROS["r_min"]
    n_colisionaron = np.sum(colisionaron)
    print(f"Particulas que alcanzaron r < {PARAMETROS['r_min']}: {n_colisionaron} ({100*n_colisionaron/n_part:.1f}%)")
```

# Idea 2: Efectos Relativistas en la Emisión de Luz del Disco

## Fundamentos Teóricos: Beaming y Corrimiento al Rojo

Aunque la dinámica de nuestras partículas se simuló bajo la aproximación newtoniana, las velocidades que alcanzan en las regiones internas del disco son fracciones significativas de la velocidad de la luz. Por tanto, la luz que emiten hacia un observador distante sufre profundas modificaciones descritas por la Relatividad Especial y General.

### 1. Corrimiento al Rojo Cinématico (Efecto Doppler Relativista)
De acuerdo a las Clases 7 y 8, la frecuencia observada $\\nu_{\\text{obs}}$ de la luz emitida por una fuente que se mueve con velocidad $\\vec{v}$ respecto a un observador está dictada por el factor Doppler relativista $\\mathcal{D}$:
$$
\\mathcal{D} = \\frac{1}{\\gamma(1 - \\vec{\\beta} \\cdot \\hat{n})}
$$
Donde:
- $\\vec{\\beta} = \\vec{v}/c$ es la velocidad normalizada.
- $\\gamma = 1/\\sqrt{1 - |\\vec{\\beta}|^2}$ es el factor de Lorentz.
- $\\hat{n}$ es el vector unitario en la línea de visión del observador distante.

### 2. Corrimiento al Rojo Gravitacional
Adicionalmente, según la escala de Schwarzschild (Clase 20), los fotones deben "escalar" el pozo de potencial gravitacional para llegar al observador en el infinito. Esto introduce una dilatación temporal gravitacional que disminuye la frecuencia observada:
$$
\\frac{\\nu_{\\text{obs}}}{\\nu_{\\text{em}}} = \\sqrt{1 - \\frac{r_s}{r}} = \\sqrt{1 - \\frac{2GM}{c^2 r}}
$$

### 3. Frecuencia Total Observada y Efecto Beaming (Faro Relativista)
Combinando ambos efectos, la relación total entre la frecuencia observada y la emitida es el factor $g$ (comúnmente usado en astrofísica de discos):
$$
g = \\frac{\\nu_{\\text{obs}}}{\\nu_{\\text{em}}} = \\mathcal{D} \\sqrt{1 - \\frac{2M}{r}}
$$
Si $g > 1$, la luz presenta un **corrimiento al azul** (blueshift), y si $g < 1$, presenta un **corrimiento al rojo** (redshift).
Debido al efecto *Beaming Relativista*, la intensidad de la radiación (flujo observado) se amplifica con el cubo de este factor ($I_{\\text{obs}} \\propto g^3 I_{\\text{em}}$), lo que hace que el lado del disco que se acerca al observador brille muchísimo más que el lado que se aleja.

### 4. Perfil de Línea Espectral (El Espectro Observado)
En la práctica astrofísica, rara vez resolvemos espacialmente el disco; en su lugar, capturamos la luz integrada de todo el sistema. Si asumimos que el gas del disco emite fuertemente en una línea espectral con frecuencia de reposo $\nu_0$ (por ejemplo, por transiciones atómicas), el observador distante detectará un espectro deformado.

La intensidad observada $I_{\text{obs}}$ a una frecuencia dada está relacionada con la intensidad emitida en el marco de reposo del gas $I_{\text{em}}$ por el teorema de invariancia del espacio de fase de los fotones (teorema de Liouville), el cual dicta que:
$$
I_{\text{obs}}(\nu_{\text{obs}}) = g^3 I_{\text{em}}(\nu_{\text{em}})
$$
Dado que la emisividad del disco suele decaer con el radio (típicamente $I_{\text{em}} \propto r^{-q}$, con $q \approx 3$ en discos de acreción estándar), los fotones originados en distintas partes del disco llegarán con diferentes corrimientos $g$. 

Al integrar la luz de todo el disco, el efecto Doppler de la materia acercándose y alejándose genera un perfil de "doble pico" ensanchado. El pico azul (blueshift) será dramáticamente más intenso que el pico rojo (redshift) debido a la amplificación del beaming relativista ($g^3$), y todo el perfil estará ligeramente desplazado hacia frecuencias más bajas debido a la penalidad del corrimiento gravitacional.

```python
# =============================================================================
# CALCULO DE EFECTOS OPTICOS RELATIVISTAS (DOPPLER + GRAVEDAD)
# =============================================================================

def graficar_emision_relativista(posiciones, velocidades, inclinacion_grados=80):
    """
    Calcula el factor de corrimiento espectral 'g' (Doppler + Gravitacional) 
    y grafica el disco de acreción visto por un observador distante.
    
    Parametros:
    - posiciones: array del estado final (n_particulas, 2) obtenido de rebound.
    - velocidades: array del estado final (n_particulas, 2).
    - inclinacion_grados: Ángulo de visión (0 = face-on / desde arriba, 90 = edge-on / de canto).
    """
    # Constantes en unidades geométricas (G=1, M=1, c=1)
    C = 1.0  
    M = 1.0  
    r_s = 2 * M # Radio de Schwarzschild
    
    # Extraemos el último estado de la simulación
    pos_finales = posiciones[-1]
    vel_finales = velocidades[-1]
    n_particulas = len(pos_finales)
    
    # Vector de la línea de visión del observador (asumiendo que está en el plano X-Z)
    inc_rad = np.radians(inclinacion_grados)
    n_obs = np.array([np.sin(inc_rad), 0.0, np.cos(inc_rad)])
    
    factores_g = np.zeros(n_particulas)
    
    for i in range(n_particulas):
        x, y = pos_finales[i]
        vx, vy = vel_finales[i]
        r = np.sqrt(x**2 + y**2)
        
        # Filtro de seguridad: si la partícula cruzó el radio de Schwarzschild, g=0 (no emite luz visible)
        if r <= r_s:
            factores_g[i] = 0.0
            continue
            
        # 1. Corrimiento gravitacional
        g_grav = np.sqrt(1 - (r_s / r))
        
        # 2. Corrimiento cinemático (Doppler Especial)
        v_vec = np.array([vx, vy, 0.0])
        beta_vec = v_vec / C
        beta2 = np.dot(beta_vec, beta_vec)
        
        # Factor de Lorentz
        gamma = 1.0 / np.sqrt(1 - beta2)
        
        # Producto punto entre velocidad y línea de visión
        beta_dot_n = np.dot(beta_vec, n_obs)
        
        # Factor Doppler relativista
        Doppler = 1.0 / (gamma * (1 - beta_dot_n))
        
        # Factor g total
        factores_g[i] = Doppler * g_grav
    # =============================================================================
    # VISUALIZACIÓN EN 3D CON PLOTLY (COLOR MAP: REDSHIFT / BLUESHIFT)
    # =============================================================================
    fig = go.Figure()

    # Añadir agujero negro (Singularidad / Horizonte)
    fig.add_trace(go.Scatter3d(
        x=[0], y=[0], z=[0],
        mode='markers',
        marker=dict(size=10, color='black'),
        name="Agujero Negro"
    ))

    # Graficar las partículas coloreadas según su factor 'g'
    fig.add_trace(go.Scatter3d(
        x=pos_finales[:, 0],
        y=pos_finales[:, 1],
        z=np.zeros(n_particulas),
        mode='markers',
        marker=dict(
            size=3,
            color=factores_g,
            colorscale='RdBu', # Rojo para redshift (se aleja), Azul para blueshift (se acerca)
            cmin=0.5,  # Límites típicos para visualizar el contraste
            cmax=1.5,
            colorbar=dict(title="Factor g<br>(v_obs / v_em)"),
            opacity=0.8
        ),
        name="Disco de Acreción"
    ))

    # Ajustar la cámara para que coincida con el ángulo de inclinación del observador
    camera = dict(
        up=dict(x=0, y=0, z=1),
        center=dict(x=0, y=0, z=0),
        eye=dict(x=2.5 * np.sin(inc_rad), y=0, z=2.5 * np.cos(inc_rad))
    )

    fig.update_layout(
        title=f"Efecto Doppler y Beaming Relativista (Inclinación: {inclinacion_grados}°)",
        scene=dict(
            xaxis_title='X (GM/c²)',
            yaxis_title='Y (GM/c²)',
            zaxis_title='Z (GM/c²)',
            aspectmode='data',
            bgcolor="rgb(20, 20, 20)", # Fondo oscuro para resaltar la luz
        ),
        scene_camera=camera,
        width=900,
        height=700,
        template="plotly_dark"
    )

    fig.show()

# Ejecutamos la función con las posiciones y velocidades obtenidas de tu bloque anterior
# Asumiendo que las variables 'posiciones' y 'velocidades' están en memoria
graficar_emision_relativista(posiciones, velocidades, inclinacion_grados=20)

```

```python
# =============================================================================
# CÁLCULO DEL ESPECTRO DE EMISIÓN (PERFIL DE LÍNEA RELATIVISTA)
# =============================================================================
import matplotlib.pyplot as plt
import numpy as np

def simular_espectro_disco(posiciones, velocidades, angulos_inclinacion=[15, 45, 80], q_emisividad=3.0):
    """
    Simula el espectro de una línea de emisión integrada sobre todo el disco
    para diferentes ángulos de observación.
    """
    C = 1.0  
    M = 1.0  
    r_s = 2 * M 
    
    pos_finales = posiciones[-1]
    vel_finales = velocidades[-1]
    n_particulas = len(pos_finales)
    
    plt.figure(figsize=(10, 6))
    
    # Colores para distinguir los ángulos
    colores = ['#2ca02c', '#ff7f0e', '#d62728'] 
    
    for idx, inc_grados in enumerate(angulos_inclinacion):
        inc_rad = np.radians(inc_grados)
        n_obs = np.array([np.sin(inc_rad), 0.0, np.cos(inc_rad)])
        
        factores_g = []
        flujos_observados = []
        
        for i in range(n_particulas):
            x, y = pos_finales[i]
            vx, vy = vel_finales[i]
            r = np.sqrt(x**2 + y**2)
            
            # Cortar emisión dentro del horizonte o muy lejos (ruido)
            if r <= r_s * 1.1:
                continue
                
            # 1. Cinemática y Gravedad (Factor g)
            g_grav = np.sqrt(1 - (r_s / r))
            beta_vec = np.array([vx, vy, 0.0]) / C
            beta2 = np.dot(beta_vec, beta_vec)
            
            if beta2 >= 1.0: # Evitar errores numéricos si v > c por el integrador clásico
                continue 
                
            gamma = 1.0 / np.sqrt(1 - beta2)
            beta_dot_n = np.dot(beta_vec, n_obs)
            Doppler = 1.0 / (gamma * (1 - beta_dot_n))
            
            g = Doppler * g_grav
            
            # 2. Intensidad y Flujo
            # Intensidad intrínseca decae con el radio r^-q
            I_em = r**(-q_emisividad) 
            
            # Amplificación por Beaming Relativista
            I_obs = (g**3) * I_em
            
            factores_g.append(g)
            flujos_observados.append(I_obs)
            
        # Generar el histograma ponderado (que actúa como nuestro espectro)
        # El eje X es el factor g (v_obs / v_em), el eje Y es el flujo total recogido
        conteos, bordes = np.histogram(factores_g, bins=60, weights=flujos_observados, density=True)
        centros_bins = (bordes[:-1] + bordes[1:]) / 2
        
        # Suavizado de la curva para mejor visualización
        plt.plot(centros_bins, conteos, label=f'Inclinación: {inc_grados}°', 
                 color=colores[idx], linewidth=2.5)
        plt.fill_between(centros_bins, conteos, alpha=0.1, color=colores[idx])

    # Línea de referencia (Frecuencia de reposo / Laboratorio)
    plt.axvline(x=1.0, color='white', linestyle='--', linewidth=1.5, label='Frecuencia de Reposo ($g=1$)')
    
    plt.title("Espectro de Emisión del Disco (Efectos Relativistas Integrados)", fontsize=14, color='white')
    plt.xlabel(r"Factor de Corrimiento $g = \nu_{\rm obs} / \nu_{\rm em}$", fontsize=12, color='white')
    plt.ylabel("Flujo Observado", fontsize=12, color='white')
    
    # Estilizado oscuro para consistencia astrofísica
    ax = plt.gca()
    ax.set_facecolor('#111111')
    fig = plt.gcf()
    fig.patch.set_facecolor('#111111')
    ax.tick_params(colors='white')
    ax.spines['bottom'].set_color('white')
    ax.spines['left'].set_color('white')
    ax.spines['top'].set_color('#111111')
    ax.spines['right'].set_color('#111111')
    
    plt.legend(facecolor='#222222', edgecolor='white', labelcolor='white')
    plt.grid(True, linestyle=':', alpha=0.3, color='white')
    plt.show()

# Ejecutamos la simulación del espectro para tres ángulos diferentes
simular_espectro_disco(posiciones, velocidades, angulos_inclinacion=[15, 45, 80])
```

# Idea 3: Dinámica Relativista en Jets Astrofísicos

## Fundamentos Teóricos: Electrodinámica Covariante y el Tensor de Faraday

Los jets astrofísicos son enormes chorros de plasma altamente colimados que emergen de los polos del agujero negro. Según vimos en la Clase 12, surgen cuando el campo magnético del disco de acreción es arrastrado y retorcido por la rotación diferencial, adoptando una configuración helicoidal a lo largo del eje de rotación (eje Z).

Para simular el movimiento de las partículas cargadas dentro de este jet, debemos abandonar la mecánica newtoniana y utilizar la **Fuerza de Lorentz explícitamente covariante** (Clase 11).

### 1. El Tensor de Faraday ($F^{\mu\nu}$)
Los campos eléctrico ($\vec{E}$) y magnético ($\vec{B}$) se unifican en un tensor antisimétrico de rango 2. En unidades donde $c=1$, el tensor con índices contravariantes es:
$$
F^{\mu\nu} = \begin{pmatrix} 
0 & E_x & E_y & E_z \\ 
-E_x & 0 & B_z & -B_y \\ 
-E_y & -B_z & 0 & B_x \\ 
-E_z & B_y & -B_x & 0 
\end{pmatrix}
$$

### 2. Campo Magnético Helicoidal
Modelaremos un campo magnético representativo del jet con una fuerte componente axial constante ($B_z = B_0$) y una componente toroidal que depende de la distancia al eje para confinar el plasma:
- $B_x = -B_1 \cdot y$
- $B_y = B_1 \cdot x$
- $B_z = B_0$

### 3. Ecuación de Movimiento Covariante
La aceleración de la partícula (cambio de su cuadrivelocidad $U^\mu$ respecto al tiempo propio $\tau$) está dictada por la contracción del tensor de Faraday con la cuadrivelocidad:
$$
\frac{dU^\mu}{d\tau} = \frac{q}{m} F^{\mu}_{\ \ \nu} U^\nu
$$
Donde $F^{\mu}_{\ \ \nu} = F^{\mu\alpha} \eta_{\alpha\nu}$, utilizando la métrica de Minkowski $\eta_{\alpha\nu} = \text{diag}(1, -1, -1, -1)$.

```python
# =============================================================================
# SIMULACIÓN DEL JET USANDO ELECTRODINÁMICA COVARIANTE
# =============================================================================
import numpy as np
from scipy.integrate import solve_ivp
import plotly.graph_objects as go

def tensor_faraday_jet(B0, B1, x, y, z):
    """
    Construye el tensor mixto de Faraday F^mu_nu para un campo helicoidal.
    """
    # Definición del campo magnético helicoidal
    Bx = -B1 * y
    By = B1 * x
    Bz = B0
    
    # Campo eléctrico (asumimos 0 en el marco del jet para este ejemplo básico)
    Ex, Ey, Ez = 0.0, 0.0, 0.0
    
    # Tensor de Faraday contravariante F^{mu nu}
    F_up_up = np.array([
        [  0,   Ex,  Ey,  Ez],
        [-Ex,   0,   Bz, -By],
        [-Ey, -Bz,   0,   Bx],
        [-Ez,  By, -Bx,   0 ]
    ])
    
    # Métrica de Minkowski eta_{alpha nu} (Signatura + - - -)
    eta_down_down = np.diag([1, -1, -1, -1])
    
    # Tensor mixto F^mu_nu = F^{mu alpha} * eta_{alpha nu}
    F_up_down = F_up_up @ eta_down_down
    return F_up_down

def edm_particula_jet(tau, Ys, q_m, B0, B1):
    """
    Sistema de ecuaciones diferenciales acopladas para la cuadriposición 
    y cuadrivelocidad de una partícula cargada.
    """
    x0, x1, x2, x3, U0, U1, U2, U3 = Ys
    
    # Cinemática: dx^mu / dtau = U^mu
    dx0_dtau = U0
    dx1_dtau = U1
    dx2_dtau = U2
    dx3_dtau = U3
    
    # Obtener el tensor de Faraday local
    F_mixto = tensor_faraday_jet(B0, B1, x1, x2, x3)
    U_vec = np.array([U0, U1, U2, U3])
    
    # Dinámica: dU^mu / dtau = (q/m) * F^mu_nu * U^nu
    dU_dtau = q_m * (F_mixto @ U_vec)
    
    return [dx0_dtau, dx1_dtau, dx2_dtau, dx3_dtau,
            dU_dtau[0], dU_dtau[1], dU_dtau[2], dU_dtau[3]]

# =============================================================================
# CONFIGURACIÓN E INTEGRACIÓN
# =============================================================================
# Parámetros del campo del Jet
B0_axial = 5.0     # Intensidad del campo magnético a lo largo de Z
B1_toroidal = 0.5  # Intensidad del enrollamiento del campo
q_m = -1.0         # Relación carga-masa (ej. electrón en unidades adaptadas)

# Condiciones iniciales de una partícula inyectada desde el disco hacia el jet
t0 = 0.0
# La inyectamos cerca del eje Z (x=1, y=0) en z=0
r_inicial = [1.0, 0.0, 0.0] 

# Le damos una velocidad inicial alta a lo largo del eje Z (v_z = 0.9c) y algo de rotación
v_inicial = [0.0, 0.1, 0.9] 
v2 = sum(v**2 for v in v_inicial)
gamma0 = 1.0 / np.sqrt(1.0 - v2)

# Cuadrivelocidad inicial U^mu = gamma * (1, v_x, v_y, v_z)
U0_inicial = gamma0
U_espacial = [gamma0 * v for v in v_inicial]

# Vector de estado inicial: [t, x, y, z, Ut, Ux, Uy, Uz]
Ys0 = [t0, r_inicial[0], r_inicial[1], r_inicial[2], 
       U0_inicial, U_espacial[0], U_espacial[1], U_espacial[2]]

# Rango de tiempo propio a simular
taus = np.linspace(0, 50, 2000)

# Integración numérica
sol_jet = solve_ivp(edm_particula_jet, 
                    (taus[0], taus[-1]), 
                    Ys0, 
                    t_eval=taus, 
                    args=(q_m, B0_axial, B1_toroidal),
                    method='RK45') # RK45 es suficiente para campos EM planos

# Extraemos la solución espacial
x_jet = sol_jet.y[1]
y_jet = sol_jet.y[2]
z_jet = sol_jet.y[3]

# =============================================================================
# VISUALIZACIÓN DEL JET 3D
# =============================================================================
fig_jet = go.Figure()

# Trayectoria de la partícula en el Jet
fig_jet.add_trace(go.Scatter3d(
    x=x_jet, y=y_jet, z=z_jet,
    mode='lines',
    line=dict(color='cyan', width=4),
    name="Trayectoria del Electrón"
))

# Representación conceptual del disco de acreción en Z=0
theta_disco = np.linspace(0, 2*np.pi, 100)
r_disco = np.linspace(2, 5, 10)
T, R = np.meshgrid(theta_disco, r_disco)
X_disco = R * np.cos(T)
Y_disco = R * np.sin(T)
Z_disco = np.zeros_like(X_disco)

fig_jet.add_trace(go.Surface(
    x=X_disco, y=Y_disco, z=Z_disco,
    colorscale='Oranges',
    opacity=0.3,
    showscale=False,
    name="Disco de Acreción"
))

fig_jet.update_layout(
    title="Confinamiento y Aceleración de Partículas en un Jet Astrofísico",
    scene=dict(
        xaxis_title='X (GM/c²)',
        yaxis_title='Y (GM/c²)',
        zaxis_title='Z (Eje del Jet)',
        bgcolor="rgb(10, 10, 15)",
    ),
    width=900, height=700,
    template="plotly_dark"
)

fig_jet.show()
```

```python
# =============================================================================
# CONFIGURACIÓN E INTEGRACIÓN: ENJAMBRE DE PARTÍCULAS
# =============================================================================
# Parámetros del campo del Jet
B0_axial = 5.0     
B1_toroidal = 0.5  
q_m = -1.0         

# Configuración del enjambre
num_particulas = 20
taus = np.linspace(0, 50, 2000)
trayectorias_x = []
trayectorias_y = []
trayectorias_z = []

np.random.seed(42) # Para reproducibilidad de los números aleatorios

for i in range(num_particulas):
    # Inyección aleatoria alrededor del borde interno del disco (ej. radio r=1.0)
    # Introducimos una pequeña dispersión espacial gaussiana
    r0_x = np.random.normal(1.0, 0.15)
    r0_y = np.random.normal(0.0, 0.15)
    r0_z = 0.0
    
    # Dispersión en la velocidad inicial (Velocidad térmica superpuesta al jet)
    # Queremos una alta velocidad media en Z, con fluctuaciones en X e Y
    v0_x = np.random.normal(0.0, 0.05)
    v0_y = np.random.normal(0.1, 0.05)
    v0_z = np.random.uniform(0.85, 0.95) # Velocidades z entre 0.85c y 0.95c
    
    v2 = v0_x**2 + v0_y**2 + v0_z**2
    
    # Control de seguridad numérico: evitar que fluctuaciones aleatorias superen 'c' (v2 >= 1.0)
    if v2 >= 0.99:
        factor = 0.99 / np.sqrt(v2)
        v0_x *= factor; v0_y *= factor; v0_z *= factor
        v2 = v0_x**2 + v0_y**2 + v0_z**2

    # Cálculo del factor de Lorentz inicial
    gamma0 = 1.0 / np.sqrt(1.0 - v2)
    
    U0_inicial = gamma0
    U_espacial = [gamma0 * v0_x, gamma0 * v0_y, gamma0 * v0_z]
    
    Ys0 = [0.0, r0_x, r0_y, r0_z, U0_inicial, U_espacial[0], U_espacial[1], U_espacial[2]]
    
    # Integración numérica para esta partícula
    sol_jet = solve_ivp(edm_particula_jet, 
                        (taus[0], taus[-1]), 
                        Ys0, 
                        t_eval=taus, 
                        args=(q_m, B0_axial, B1_toroidal),
                        method='RK45')
    
    # Guardamos las coordenadas espaciales
    trayectorias_x.append(sol_jet.y[1])
    trayectorias_y.append(sol_jet.y[2])
    trayectorias_z.append(sol_jet.y[3])

# =============================================================================
# VISUALIZACIÓN DEL JET 3D (MÚLTIPLES TRAYECTORIAS)
# =============================================================================
fig_jet = go.Figure()

# Paleta de colores para dar efecto de "haz" de plasma
colores_plasma = ['#00ffff', '#00bfff', '#1e90ff', '#4169e1', '#0000ff']

# Graficar cada partícula
for i in range(num_particulas):
    color = colores_plasma[i % len(colores_plasma)]
    fig_jet.add_trace(go.Scatter3d(
        x=trayectorias_x[i], y=trayectorias_y[i], z=trayectorias_z[i],
        mode='lines',
        line=dict(color=color, width=2.5),
        opacity=0.7,
        showlegend=False
    ))

# Representación conceptual del disco de acreción en Z=0
theta_disco = np.linspace(0, 2*np.pi, 100)
r_disco = np.linspace(2, 5, 10)
T, R = np.meshgrid(theta_disco, r_disco)
X_disco = R * np.cos(T)
Y_disco = R * np.sin(T)
Z_disco = np.zeros_like(X_disco)

fig_jet.add_trace(go.Surface(
    x=X_disco, y=Y_disco, z=Z_disco,
    colorscale='Oranges',
    opacity=0.3,
    showscale=False,
    name="Disco de Acreción"
))

# Configuración visual
fig_jet.update_layout(
    title="Estructura Macroscópica del Jet: Confinamiento de un Enjambre de Electrones",
    scene=dict(
        xaxis_title='X (GM/c²)',
        yaxis_title='Y (GM/c²)',
        zaxis_title='Z (Eje del Jet)',
        bgcolor="rgb(10, 10, 15)",
        # Fijar los límites para que la escala no se desajuste
        xaxis=dict(range=[-3, 3]),
        yaxis=dict(range=[-3, 3])
    ),
    width=900, height=700,
    template="plotly_dark"
)

fig_jet.show()
```
